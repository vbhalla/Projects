```python
#let's remove features that have a lot of nulls
def count_null_df(df):
    feature_null_perc = {}
    for column in df: 
        null_sum = float(df[column].isnull().sum())
        if null_sum > 0:
            tot_col = len(df[column]) 
            null_perc = (null_sum/tot_col)*100  
            feature_null_perc[column] = null_sum, null_perc
    return feature_null_perc
```

```python
f_null = count_null_df(df_2014)
f_null = pd.DataFrame(f_null).transpose()
f_null.sort_values(by=1, axis=0,ascending=False, inplace=True)
f_null.rename(columns = {0:"num_nulls", 1:"perc_nulls"}, inplace = True)
```

```python
#set the default flags based off of loan status
def default_label(row):
    if row == "Current":
        return 0
    elif row == "Fully Paid":
        return 0
    elif row == "Charged Off":
        return 1
    elif row == "Default":
        return 1
    elif row == "Late (31-120 days)":
        return 1
    elif row == "Late (16-30 days)":
        return -1
    elif row == "In Grace Period":
        return -1
    else:
        return -1000

df_reduce['loan_label'] = df_reduce.loan_status.apply(lambda x: default_label(x))

df_reduce['loan_label'] = df_reduce.loan_status.apply(lambda x: default_label(x))
```
